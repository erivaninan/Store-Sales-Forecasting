## Comparaison entre la prevision global et la bottom up

```python
import os
import pandas as pd
import plotly.graph_objects as go
from functions import *
import plotly.express as px
```

```python
# --recuperer les ventes du testing--

parent_directory = os.path.abspath(os.path.join(os.getcwd(), ".."))
models_directory = os.path.join(parent_directory, "models")
testing_path = os.path.join(models_directory, "series_test.csv")
test = pd.read_csv(testing_path)
```

```python
test.head()
```

```python
Cluster_0 = pd.read_csv('Cluster_0.csv')
Cluster_1 = pd.read_csv('Cluster_1.csv')
Cluster_2 = pd.read_csv('Cluster_2.csv')
Cluster_3 = pd.read_csv('Cluster_3.csv')
Cluster_4 = pd.read_csv('Cluster_4.csv')

All_prevision = Cluster_0
All_prevision['sales'] = test['sales']
All_prevision['test_Forecast_prophet_online_global'] = test['test_Forecast_prophet_online']
All_prevision['Test_Forecast_Sarimax_global'] = test['Test_Forecast_Sarimax']
```

```python
All_prevision = All_prevision .drop(columns=['cluster'])
All_prevision = All_prevision .drop(columns=['Test Forecast'])
All_prevision['Test_Forecast'] += Cluster_1['Test_Forecast'] + Cluster_2['Test_Forecast'] + Cluster_3['Test_Forecast'] + Cluster_4['Test_Forecast']
All_prevision['Test_Forecast_prophet_online'] = Cluster_0['test_Forecast_prophet_online'] + Cluster_1['test_Forecast_prophet_online'] + Cluster_2['test_Forecast_prophet_online'] + Cluster_3['test_Forecast_prophet_online'] + Cluster_4['test_Forecast_prophet_online']
All_prevision.head()
```

```python
fig = px.line(All_prevision, x='date', y=['sales', 'Test_Forecast','Test_Forecast_Sarimax_global'],title="Predictions du test Sarima")
fig.show()
```

```python
fig = px.line(All_prevision, x='date', y=['sales', 'Test_Forecast_prophet_online','test_Forecast_prophet_online_global'],title="Predictions du test Prophet Online")
fig.show()
```

bottom up en sarima c'est pas fou , prophetne ligne tres bien aussi. on a comaprer avec le meilleurs et le pire model.
